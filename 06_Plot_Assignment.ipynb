### Faceting
- ggplot has a special technique called faceting that allows to split one plot into multiple plots based on a factor included in the dataset. We will use it to make one plot for a time series for each species.

```python
import numpy as np
import pandas as pd
from plotnine import *

%matplotlib inline
my_file=pd.read_csv("C:\\Users\\1233w\\Desktop\\Assignment\\Chilla_data2_for_plots.csv")
ggplot(my_file, aes(x="Gender", y="Height in cm? Freelancer- (Float)",size = "Age (years)-Float/Int",)) + geom_point()
```

```python
import numpy as np
import pandas as pd
from plotnine import *

%matplotlib inline
my_file=pd.read_csv("C:\\Users\\1233w\\Desktop\\Assignment\\Chilla_data2_for_plots.csv")
ggplot(my_file, aes(x="Gender", y="Height in cm? Freelancer- (Float)",size = "Age (years)-Float/Int",
                    color = "Marital Status?")) + geom_point()
```

```python
import numpy as np
import pandas as pd
from plotnine import *

%matplotlib inline
my_file=pd.read_csv("C:\\Users\\1233w\\Desktop\\Assignment\\Chilla_data2_for_plots.csv")
ggplot(my_file, aes(x="Gender", y="Height in cm? Freelancer- (Float)",size = "Age (years)-Float/Int",
                    )) + geom_point() + facet_wrap(["Marital Status?"]) +    theme(axis_text_x = element_text(size=6))


```
