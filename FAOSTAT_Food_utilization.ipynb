# 04.Box plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

child_malnutrition=pd.read_csv("C:\\Users\\1233w\\Desktop\\FAOSTAT_child malnutrition.csv")

sns.boxplot(x="Item Code",y="Year",showmeans=True,
 meanprops= {"marker":"*",
 "markersize":"12",
 "markeredgecolor":"red"},
 data=child_malnutrition)                 
        
plt.xlabel("Child Health",size=10),
plt.ylabel("Age (years)",size=10),
plt.title("Age Make Long",size=10,weight="bold")
```
